```python
class Queue:
  def __init__(self):
    self.queue = []

  def enqueue(self,item):
    self.queue.append(item)

  def dequeue(self):
    if len(self.queue) == 0:
      return None
    else:
      return self.queue.pop(0)
    
  def __len__(self):
    return len(self.queue)
```

```python

import networkx as nx

map = {'A':'TZS','B':'PGUF','C':'DRP','D':'MC','E':'H','F':'SB','G':'B','H':'UE','I':'NV','L':'TM','M':'LD','N':'I','O':'ZS','P':'RBC','R':'PCS','S':'RFAO','T':'LA','U':'HVB','V':'IU','Z':'AO'}

g = nx.DiGraph()

edge_list = []
for s in map.keys():
  for c in map[s]:
    edge_list.append((s,c))
g.add_edges_from(edge_list)
nx.draw_networkx(g)

startNode = input('Input your start node : ').upper()
goalNode = input('Input your goal node : ').upper()
breadthFirstSearch(startNode,goalNode)
```

```python
def breadthFirstSearch(start,goal):
  wait = Queue()
  explored = []
  frontier = Queue()
  if start == goal:
    return startNode
  wait.enqueue(start)
  explored.append(start)
  while len(set(wait.queue))< len(map.keys()):
    
    while len(wait) > 0:
      node = wait.dequeue()
      for i in map[node]:
        if not i in explored:
          frontier.enqueue(i)

    while len(frontier.queue) > 0:
      state = frontier.dequeue()
      explored.append(state)
      if state == goal:
        return explored
      else:

        wait.enqueue(state)
    
      
      
  
    
    
  
```
